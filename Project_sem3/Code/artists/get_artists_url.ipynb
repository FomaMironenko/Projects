## Getting URLs for artists pages from mutalart.com

#### Initially inpath (artists.csv) contains a table with artists names

```python
import sys
try:
    sys.path.insert(0, "/usr/lib/python3.7/site-packages")
except FileNotFoundError:
    pass

import numpy as np
import pandas as pd

import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
```

### Iput:      artist name
### Output: an adress of the page at mutalart.com

```python
def get_url(name):
    print(name)
    domain = "https://www.mutualart.com/"
    URL = domain + "Artist/" + name.strip().rstrip().replace(" ", "-")
    agent = UserAgent()
    agent.chrome
    page = requests.get(URL, headers={'User-Agent': agent.chrome})
    text = BeautifulSoup(page.text, "html.parser")
    all_options = text.find_all(class_="item col-xs-3")
    
    if(len(all_options) == 0):
        return ""
    link   = all_options[0].find(class_="thumbnail")
    adress = str(link)
    l = adress.find("href=")
    r = adress.find(">", l)
    adress = adress[l+7:r-1]
    URL = domain + adress
    return URL
```

```python
inpath = "artists.csv"
df = pd.read_csv(inpath)
del(df["Unnamed: 0"])
df
```

```python
df["URL"] = [get_url(x) for x in df.Artists]
```

```python
df.to_csv(inpath)
```
