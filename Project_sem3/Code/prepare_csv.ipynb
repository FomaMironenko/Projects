```python
import numpy as np
import pandas as pd
```

```python
inpath = "painters"
text = open(inpath).read()
```

```python
items = text.split("\n")
while(True): 
    try:
        items.remove("")
    except ValueError:
        break
items = items[6:]
for item in items:
    item = item.replace("\t", "")
butches = [items[i:i+7] for i in range(0, len(items), 5)]

df = pd.DataFrame(columns=["Artist", "Name", "date", "price", "comments", "repeat_sale"])
    
df.Artist = [x[1].strip().rstrip() for x in butches]
df.repeat_sale = [int("REPEAT SALE" in x[2]) for x in butches]
df.Name = [x[2] for x in butches]
#replacing REPEAT SALE
df.Name = df.Name.apply(lambda x: x.replace("REPEAT SALE", ""))

df.date = [x[3] for x in butches]
df.price = [x[4] for x in butches]
df.comments = [x[2][x[2].find("("):].replace("(", "").replace(")", "").replace("REPEAT SALE", "") for x in butches]
#replacing comments
df.Name = df.Name.apply(lambda x: x[:x.find("(")].strip().rstrip())
```
